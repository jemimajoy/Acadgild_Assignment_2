```python
import pandas as pd
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",skipinitialspace = True)
df.loc[-1]=df.columns.tolist()
df.index = df.index+1
df = df.sort_index()
df.columns =['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','50k']
#df.to_csv("C:\\Users\\PACKIJO\\Desktop\\acad\\adult.csv")
df.head()

```

```python
from sqlalchemy import create_engine
engine = create_engine("sqlite://", echo = False)
df.to_sql('new_adult', con = engine, if_exists='replace')
engine.execute("select new_adult.'native-country' from new_adult where workclass = 'Private'").fetchall()
```

```python
#Write two basic update queries
engine.execute("""UPDATE new_adult
SET workclass = 'Unknown'
WHERE workclass = '?'""")
print(engine.execute("select workclass from new_adult where workclass= '?'").fetchall())
```

```python
engine.execute(""" UPDATE new_adult 
                SET sex ='M' 
                WHERE sex='Male'""")
engine.execute("select sex from new_adult").fetchall()
```

```python
#Write two delete queries
engine.execute("""DELETE  FROM new_adult
where new_adult.age <= 30 """)
engine.execute("""select new_adult.age from new_adult """).fetchall()
```

```python
engine.execute("""DELETE FROM new_adult
WHERE new_adult.'education' = '11th'""")
engine.execute("""select * from new_adult where new_adult.'education' = '11th' """).fetchall()
```

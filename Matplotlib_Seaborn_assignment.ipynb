```python
#Create a pie chart presenting the male/female proportion
pop =titanic['sex'].value_counts().tolist()
plt.pie(pop,labels = ['Female','Male'],startangle=90,autopct= '%2.2f%%', shadow = True, colors = ['b', 'g'])
plt.title("Titanic")
plt.legend()
```

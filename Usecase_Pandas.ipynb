```python
#Find how many null values, missing values are present. Deal with them. Print out how many rows have been modified.
import numpy as np
#df.isnull().sum().sum()
df.isnull().count()
```

```python
#Filter out movies from the movies DataFrame that are of type 'Animation'
df2[df2['genres'].str.contains("Animation")]
```

```python
#Find the average rating of movies.
df.groupby("movieId")["rating"].agg(np.mean).head()
```

```python
#Perform an inner join of movies and tags based on movieId.
df2.join(df1.set_index('movieId'), on='movieId').head()
```

```python
#Print out the 5 movies that belong to the Comedy genre and have rating greater than 4.
df2[df2['movieId'].isin(df.loc[df['movieId'].isin(df2[df2['genres'].str.contains('Comedy')]['movieId'])& (df['rating'] >= 4.0)]['movieId'].head())]['title']
```

```python
#Split 'genres' into multiple columns.
df2['genres'].str.get_dummies('|').replace({0:'no', 1:'yes'}).head()
```

```python
#Extract year from title e.g. (1995).
df2['title'].str.extract(r'([0-9]+)')[0].head()
```

```python
#Select rows based on timestamps later than 2015-02-01.
from datetime import datetime
df1['timestamp1'] = df1['timestamp'].apply(datetime.fromtimestamp)
print(df1[df1['timestamp1']> '2015-02-01'].head())
```

```python
#Sort the tags DataFrame based on timestamp
df1.sort_values('timestamp')['tag'].head()
```
